```python
cat = []
num = []

for x in data.columns:
    if data.dtypes[x]=='object':
        cat.append(x)
    else:
        num.append(x)
```

```python
for x in data.columns:
    nulls = data[x].isnull().sum()
    if nulls != 0:
        print(x, ': ', nulls, sep='')
        if x in cat:
            print(data[x].unique())
        print('\n')
print('\ntotal', ': ', data.shape[0], sep='')
```
